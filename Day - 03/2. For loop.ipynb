## range()

```python
range(10) -> 0 to 9
range(3, 10) -> 3 to 9
range(2, 20, 4) -> 2 to 19 and third arg 4 is step_size 
```

```python
for i in range(10):
    print(i, end = ' ')
```

```python
for xyz in range(3,23):
    print(xyz, end = ' ')
```

```python
for i in range(6):
    for j in range(6):
        print(i,j, sep = ' ')
```

```python
for i in range(6):
    for j in range(6):
        print(j,i, sep = ' ')
```

```python
for i in range(6):
    for j in range(6):
        print('(', i, j, ')', end = ',', sep = '')
    print()
```

```python
for i in range(1,51):
    if i%2 == 0:
        print(i, end = ' ')
```

```python
for i in range(1,51):
    if i%2 != 0:
        print(i, end = ' ')
```
